```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

### 1. 역대 박스오피스(연도별) 영화 이름, 매출, 관객정보 크롤링

```python
url = "http://www.kobis.or.kr/kobis/business/stat/boxs/findYearlyBoxOfficeList.do"
years = list(range(2004,2020))
info = []
for year in years :
    par={
        "CSRFToken":"vvd9gl6efm3vZOKAsJoK7YY7Q3lIhT4vmL4Iof_4ovE",
        "loadEnd": 0 ,
        "searchType":"search" ,
        "sSearchYearFrom": year}
    response = requests.post(url,par)
    dom = BeautifulSoup(response.content.decode('utf-8'), "html.parser")
    elements = dom.select("#td_movie > span > a")
    for i in range(0,50):
        title = dom.select("#td_movie > span > a")[i].text
        sales = dom.select("#td_salesAcc")[i].text.strip()
        audience = dom.select("#td_audiAcc")[i].text.strip()
        date = dom.select("#td_openDt")[i].text.strip()
        info.append({"title":title,"sales":sales,"audience":audience,"date":date})
        print(title,date)
```

```python
info_df = pd.DataFrame(info)
info_df["date"]=pd.to_datetime(info_df['date'])
info_df["year"]=info_df["date"].dt.year
info_df.tail(2)
```

```python
info_df[info_df["title"].str.contains("&")]
```

```python
info_df = info_df.drop([58,317,763]).reset_index()
info_df.tail(2)
```

### 2.영화 코드 크롤링

```python
url = "http://www.kobis.or.kr/kobis/business/mast/mvie/searchUserMovCdList.do"
movie_code=[]
for title,year in zip(list(info_df["title"]),list(info_df["year"])):
        par={
        "SRFToken":"C4_4vSS5xYH_1JtsmQhX3XCkerqxmcO3SXGnG4vJmdQ",
        "curPage": 1,
        "ordering":"updDttmOrder",
        "searchOpen":"Y",
        "movieNm":title,
        "openStartDt":year,
        "openEndDt":year,
                        }
        response = requests.post(url,par)
        dom = BeautifulSoup(response.content, "html.parser")
        result = dom.select("#content > div.rst_sch > table > tbody > tr:nth-child(1) > td.last-child")[0].text
        movie_code.append({"title":title,"code":result})
        print(title,result)
```

```python
code_df = pd.DataFrame(movie_code)
code_df.tail(2)
```

### 3.영화 정보 크롤링

```python
url="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"
result_list=[]
for cd in list(code_df["code"]):
    par={
        "key":"9b748f2fadd5ced8ea2a349549d8fc6e",
        "movieCd":cd}
    response = requests.get(url,par)
    result = response.json()["movieInfoResult"]["movieInfo"]
    result_list.append(result)
    print(cd)
```

```python
df = pd.DataFrame(result_list)
df.head(5)
```
