### 2. 전처리

```python
def clean(row):
    result = str(row).replace("{"," ").replace("]"," ").replace("}"," ").replace("'"," ").replace(":"," ").replace("["," ").replace("("," ").replace(")"," ").replace("-"," ")
    return result
def striptext(row):
    result = str(row).strip()
    result = str(row).replace(" ","")
    return result
def clean2(row):
    result = str(row).split(",")
    text =""
    for i in range(0,len(result)):
        if i % 2 == 0:
            text = text+(str(row).split(",")[i])
    return text
```

```python
columns=["movieCd","movieNm","movieNmEn","genres","nations","openDt","prdtStatNm","prdtYear","typeNm","showTm","sales","audience","directors","actors","audits","companys","staffs"]
df = df[columns]
```

```python
df['actors'] = df['actors'].apply(clean)
df['audits'] = df['audits'].apply(clean)
df['companys'] = df['companys'].apply(clean)
df['actors'] = df['actors'].apply(clean)
df['directors'] = df['directors'].apply(clean)
df['genres'] = df['genres'].apply(clean)
df['nations'] = df['nations'].apply(clean)
df['staffs'] = df['staffs'].apply(clean)
df['movieNm'] = df['movieNm'].apply(clean)
df['movieNmEn'] = df['movieNmEn'].apply(clean)
df['sales'] = df['sales'].apply(clean)
df['audience'] = df['audience'].apply(clean)
df.tail(1)
```

```python
df['actors'] = df['actors'].apply(striptext)
df['audits'] = df['audits'].apply(striptext)
df['companys'] = df['companys'].apply(striptext)
df['actors'] = df['actors'].apply(striptext)
df['directors'] = df['directors'].apply(striptext)
df['genres'] = df['genres'].apply(striptext)
df['nations'] = df['nations'].apply(striptext)
df['staffs'] = df['staffs'].apply(striptext)
df.tail(1)
```

```python
df["genres"] = df["genres"].str.replace("genreNm","")
df["nations"] = df["nations"].str.replace("nationNm","")
df["directors"] = df["directors"].str.replace("peopleNm","").str.replace("peopleNm","peopleNmEn")
df["actors"] = df["actors"].str.replace("peopleNm","").str.replace("castEn","").str.replace("cast","")
df["audits"] = df["audits"].str.replace("watchGradeNm","")
df["companys"] = df["companys"].str.replace("companyNm","").str.replace("companyNmEn","")
df["staffs"] = df["staffs"].str.replace("peopleNm","").str.replace("peopleNm","peopleNmEn")
df.head(5)
```

```python
df["sales"] = df["sales"].str.replace(",","")
df["audience"] = df["audience"].str.replace(",","")
```

```python
df["sales"] = df["sales"].astype("int64")
df["audience"] = df["audience"].astype("int64")
```

#### 5. 러닝타임

```python
df.loc[df["showTm"]=="","showTm"] = 0
```

```python
df["showTm"] = df["showTm"].astype("int64")
```

```python
df.loc[(df["showTm"]<90) & (df["showTm"]>0),"time"] = "90분이하"
df.loc[(df["showTm"]<100) & (df["showTm"]>=90),"time"] = "90~100분"
df.loc[(df["showTm"]<110) & (df["showTm"]>=100),"time"] = "100~110분"
df.loc[(df["showTm"]<120) & (df["showTm"]>=110),"time"] = "110분~120분"
df.loc[(df["showTm"]<130) & (df["showTm"]>=120),"time"] = "120분~130분"
df.loc[(df["showTm"]<140) & (df["showTm"]>=130),"time"] = "130분~140분"
df.loc[(df["showTm"]>=140),"time"] = "140분이상"
```

```python
time = pd.pivot_table(data=df,index="time",values="count",aggfunc=sum).sort_values(by="count").reset_index()
time
```

```python
plt.figure(figsize=(12,5))
plt.title("러닝타임별 영화수")
sns.barplot(data=time,x="time",y="count")
plt.show()
```
